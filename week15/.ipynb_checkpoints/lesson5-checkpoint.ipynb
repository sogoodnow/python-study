```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=['Q1','Q2','Q3','Q4']
y=[10,30,20,60]
plt.bar(x,y)
```

```python
plt.bar(x,y,color='r',width=0.3)
plt.grid(True)
plt.text(0.2,50,'test')
```

```python
rect=plt.bar(x,y,color='r',width=0.3)
plt.grid(True)
plt.ylim(0,70)
#plt.text(0.2,50,'test')
for ind, item in enumerate(rect):
    _x = item.get_x()
    _y = item.get_height()
    plt.text(_x,_y,y[ind])
```
