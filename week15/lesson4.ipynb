```python
x=np.linspace(0,10,100)
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
y2 = np.cos(x)
```

```python
plt.plot(x,y)
plt.plot(x,y2)
```

```python
plt.plot(x,y,label='sinx')
plt.plot(x,y2,label='cosx')
plt.xlabel("x zhou")
plt.legend()
```
