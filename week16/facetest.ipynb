```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('lena.png')
```

```python
# cv.imshow('aa',img)
# cv.waitKey(0)
# cv.destroyAllWindows()
```

```python
# plt.imshow(img) #RGB BGR
# plt.show()
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB)) #RGB BGR
plt.show()
```

```python
img_gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
clf = cv.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
objects = clf.detectMultiScale(img_gray)  # 
```

```python
objects
```

```python
img2 = img.copy()
```

```python
for (x,y,w,h) in objects:
    cv.rectangle(img2,(x,y),(x+w,y+h),color=(255,0,0),thickness=2)
plt.imshow(cv.cvtColor(img2,cv.COLOR_BGR2RGB)) #RGB BGR
plt.show()
```

```python
clf = cv.CascadeClassifier('haarcascade_eye.xml')
```

```python
objects = clf.detectMultiScale(img_gray)  # 
objects
```

```python
img2 = img.copy()
for (x,y,w,h) in objects:
    cv.rectangle(img2,(x,y),(x+w,y+h),color=(255,0,0),thickness=2)
plt.imshow(cv.cvtColor(img2,cv.COLOR_BGR2RGB)) #RGB BGR
plt.show()
```
