```python
data={
    'time':pd.date_range('2018-06-01',periods=2000000,freq='T'),
    'cpu':np.random.randn(2000000)+10
}
```

```python
df=pd.DataFrame(data,columns=['time','cpu'])
df.head()
```

```python
df2=df.set_index("time")
```
